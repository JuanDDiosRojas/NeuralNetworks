# Resolviendo ODE- PDE con deep learning
Vamos a resolver la ecuación diferencial 
$$\frac{du}{dt}=2t \equiv F(t,u)$$

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 Definir la estructura de la red
ANN = nn.Sequential(nn.Linear(1, 50), nn.Sigmoid(), nn.Linear(50,1, bias=False))
```

```python
#2 Definir la función de la ecuación diferencial F
F = lambda x: 2*x
```

```python
#3 Se establece la función de coste
def coste(x):
    x.requires_grad=True
    #calculamos una salida de la red
    salida=ANN(x)
    #calculamos la derivada de la red respecto a esa entrada y salida
    ANN_t = torch.autograd.grad(salida, x, grad_outputs=torch.ones_like(salida),create_graph=True)
    #return torch.linalg.norm((ANN_t[0]-F(x)))**2
    return torch.linalg.norm((ANN_t[0]-F(x)))**2 + torch.linalg.norm((ANN(torch.tensor([0])-torch.zeros_like(x))))**2
```

```python
#4 Error y optimizador
learning_rate=0.1
epocas=1000

optimizer=torch.optim.Adam(ANN.parameters(),lr=learning_rate)
```

```python
#5 el intervalo que queremos para x
x = torch.linspace(0,10,100)
x=x.reshape(100,1)
```

```python
#loop de entrenamiendo 
errores=[]
for i in range(epocas):
    l=coste(x) #coste
    l.backward() #gradiente
    optimizer.step() #se actualizan los parámetros
    optimizer.zero_grad() #vacíamos el gradiente
    print('época=%.2f , loss=%.2f'%(i,l))
    errores.append(float(l))
plt.plot(range(epocas),errores)
```

```python
x = torch.linspace(0,10,100)
x=x.reshape(100,1)
plt.plot(x,x**2,'ro')
plt.plot(x,ANN(x).detach().numpy(),'b*')
```

# Ejemplo 2
Usando las condiciones iniciales al definir una función auxiliar de la forma:
$$u_0 + (t-a)N(t)$$
donde $u_0$ es la condición inicial de la función $u$ en el valor a: $u(a)=u_0$

$$\frac{du}{dt}=e^{-\frac{x}{5}} cos(x)-\frac{1}{5}u$$

Para resolver $$\frac{du}{dt}=F(u,t)$$

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 Definir la estructura de la red
ANN = nn.Sequential(nn.Linear(1, 50), nn.Sigmoid(), nn.Linear(50,1, bias=False))
```

```python
#2 Establecemos las condiciones iniciales
a=torch.tensor([0.0])
u0=torch.tensor([0.0])
```

```python
#3 Escribimos la ecuación que define la ecuación diferencial 
F = lambda t, u: torch.exp(-t / 5.0) * torch.cos(t) - u / 5.0
```

```python
#3 Se establece la función de coste
def coste(t):
    t.requires_grad=True
    #calculamos una salida de la red
    salida=ANN(t)
    #calculamos la derivada de la red respecto a esa entrada y salida
    ANN_t = torch.autograd.grad(salida, t, grad_outputs=torch.ones_like(salida),create_graph=True)
    #return torch.linalg.norm((ANN_t[0]-F(x)))**2
    #return torch.linalg.norm((ANN_t[0]-F(x,salida)))**2
    return  torch.mean(( ANN_t[0] - F(t,salida))** 2) + (ANN(a)-u0)** 2

```

```python
#5 el intervalo que queremos para x
x = torch.linspace(0,5,100)
x=x.reshape(100,1)
```

```python
#loop de entrenamiendo
learning_rate=0.1
epocas=500

#4 El optimizador
optimizer=torch.optim.Adam(ANN.parameters(),lr=learning_rate)

errores=[]
for i in range(epocas):
    l=coste(x) #coste
    l.backward() #gradiente
    optimizer.step() #se actualizan los parámetros
    optimizer.zero_grad() #vacíamos el gradiente
    print('época=%.2f , loss=%.5f'%(i,l))
    errores.append(float(l))
plt.plot(range(epocas),errores)
```

```python
x = torch.linspace(0,5,100)
x=x.reshape(100,1)
#plt.plot(x,2*torch.arctan(1.83*torch.exp(-torch.sin(x))),'r')
plt.plot(x,ANN(x).detach().numpy(),'r*')
plt.plot(x,torch.exp(-x / 5.0) * torch.sin(x),c='darkblue')
```

# Sistemas de ecuaciones 

Vamos a suponer que tenemos un sistema que puede ser escrito de la forma:

$$\frac{du_1}{dt}=F_1(t; \:u_1,u_2...u_n)$$


$$\frac{du_2}{dt}=F_2(t; \:u_1,u_2...u_n)$$
$$\vdots$$
$$\frac{du_n}{dt}=F_n(t; \:u_1,u_2...u_n)$$

La idea aquí, como siempre, será definir una red de:  $t\in [a,b]\subset R \longrightarrow R^n$ 

Donde la salida de la red corresponderá a cada una de las soluciones del sistema: es decir 

$$ANN(t)= \begin{pmatrix} F_1\\F_2\\ \vdots \\ F_n \end{pmatrix}$$


Por otro lado, la ecuación de coste estará definida por la suma de las diferencias entre la salida de la red y las condiciones iniciales correspondiente a cada solución. 

Así $C$ debería pasar a definirse como:
$$C=\sum_i^n \left( \frac{d}{dt}ANN_i-F_i \right) + \sum_i^n\left(ANN(a)_i-u_{i,0}\right)$$

Donde El sistema tiene condiciones iniciales $u_i(a)=u_{i,0}$

Primero hay que tener definido el sistema, para este caso usaremos:

\begin{align} 
\frac{du_1}{dt}&=u_1\\
\frac{du_2}{dt}&=u_1+u_2
\end{align}

con $u_{1,0}=u_1(0)=1$ y $u_{2,0}=u_2(0)=2$

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 Definir la estructura de la red, en este caso como son dos soluciones
#la red debe tener dos elementos en su salida
ANN = nn.Sequential(nn.Linear(1, 100), nn.Sigmoid(), nn.Linear(100, 100),nn.Sigmoid(), nn.Linear(100, 100), nn.Sigmoid(),nn.Linear(100,2, bias=False))
```

```python
#2 Establecemos las condiciones iniciales

a=torch.tensor([0.0])#valor donde se evaluan las u_i
u10=torch.tensor([1.0])#valor de u_1(a)
u20=torch.tensor([2.0])#valor de u_2(a)
```

```python
F1 = lambda u1: u1
F2 = lambda u1,u2: u1+u2
```

```python
def coste(t):
    t.requires_grad=True
    #calculamos una salida de la red, recordando que ANN[0]=u1, ANN[1]=u2
    salida1=ANN(t)[:,0]
    salida2=ANN(t)[:,1]
    #calculamos la derivada de la red respecto a esa entrada y salida
    ANN1_t = torch.autograd.grad(salida1, t, grad_outputs=torch.ones_like(salida1),create_graph=True) #ANN[0]'=u1'
    ANN2_t = torch.autograd.grad(salida2, t, grad_outputs=torch.ones_like(salida2),create_graph=True) #ANN[1]'=u2'
    
    #Ahora hacemos las dif entre cada derivada y su funcion
    c = torch.mean(( ANN1_t[0] - F1(salida1))** 2) + torch.mean(( ANN2_t[0] - F2(salida1,salida2))** 2) + (ANN(a)[0]-u10)**2 + (ANN(a)[1]-u20)**2
    return c
```

```python
#5 el intervalo que queremos para x
x = torch.linspace(0,5,100)
x=x.reshape(100,1)
```

```python
#loop de entrenamiendo
learning_rate=0.01
epocas=500

#4 El optimizador
optimizer=torch.optim.Adam(ANN.parameters(),lr=learning_rate)

errores=[]
for i in range(epocas):
    l=coste(x) #coste
    l.backward() #gradiente
    optimizer.step() #se actualizan los parámetros
    optimizer.zero_grad() #vacíamos el gradiente
    print('época=%.2f , loss=%.5f'%(i,l))
    errores.append(float(l))
plt.plot(range(epocas),errores)
```

```python
x = torch.linspace(0,5,100)
x=x.reshape(100,1)
#plt.plot(x,2*torch.arctan(1.83*torch.exp(-torch.sin(x))),'r')
plt.plot(x,ANN(x)[:,0].detach().numpy(),'green',label='$u_1$')
plt.plot(x,ANN(x)[:,1].detach().numpy(),'darkblue',label='$u_2$')

#plt.plot(x,torch.exp(x),label='sol 1 real')
#plt.plot(x,0.5*torch.exp(x)+3/2*torch.exp(-x),label='sol 2 real')
plt.legend()
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from torch.autograd.functional import jacobian
```

```python
#1 Definir la estructura de la red
nodos=40
ANN = nn.Sequential(nn.Linear(2, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos,1,bias=False)
                   )
```

```python
print(ANN)
```

```python
#2 Escribimos la ecuación que define la ecuación diferencial 
F = lambda u,ut,ux,uxx: ut + u*ux - (0.01/3.14159)*uxx
```

```python
#3 Se establece la función de coste
def coste(t):
    t.requires_grad=True
    #calculamos una salida de la red
    salida=ANN(t)
    #calculamos la derivada de la red respecto a esa entrada y salida, recordemos que cada elemento
    #de la función grad(), es la derivada parcial respecto a cada  una de las entradas
    DANN = torch.autograd.grad(salida, t, grad_outputs=torch.ones_like(salida),create_graph=True)[0] #Gradiente
    ANN_t=DANN[:,0] #dANN/dt
    ANN_x=DANN[:,1] #dANN/dx
    
    #Ahora calculamos las derivadas de segundo orden
    DDANN = torch.autograd.grad(DANN, t, grad_outputs=torch.ones_like(DANN),create_graph=True)[0]
    ANN_tt=DANN[:,0] #dANN/dt
    ANN_xx=DANN[:,1] #dANN/dx
    
    a=torch.cat((t[:,0].reshape(t.shape[0],1),-torch.ones(t.shape[0],1)),axis=1)
    b=torch.cat((t[:,0].reshape(t.shape[0],1),torch.ones(t.shape[0],1)),axis=1)
    c=torch.cat((torch.zeros(t.shape[0],1),t[:,1].reshape(t.shape[0],1),),axis=1)
    

    return  torch.mean(F(salida,ANN_t,ANN_x,ANN_xx)** 2) + torch.mean(ANN(a)**2) + torch.mean(ANN(b)**2) + torch.mean((ANN(c)-torch.sin(-3.1415*t[:,1].reshape(t.shape[0],1)))**2)
```

```python
#5 el intervalo que queremos para x
x = torch.arange(-1,1.1,0.05)
t = torch.arange(0,1.1,0.05)
X=torch.cartesian_prod(t,x)
X.requires_grad=True
```

```python
#loop de entrenamiendo
learning_rate=0.01
epocas=1000

#4 El optimizador
optimizer=torch.optim.Adam(ANN.parameters(),lr=learning_rate)

errores=[]
for i in range(epocas):
    l=coste(X) #coste
    l.backward() #gradiente
    optimizer.step() #se actualizan los parámetros
    optimizer.zero_grad() #vacíamos el gradiente
    print('época=%.2f , loss=%.5f'%(i,l))
    errores.append(float(l))
plt.plot(range(epocas),errores)
```

## Modelando la ecuación de onda

$$\frac{\partial^2 u}{\partial t^2}-c\frac{\partial^2 u}{\partial x^2}$$

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from torch.autograd.functional import jacobian
import time
```

```python
#1 Definir la estructura de la red
nodos=30
ANN = nn.Sequential(nn.Linear(2, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos, nodos), nn.Tanh(),
                    nn.Linear(nodos,1,bias=False)
                   )
```

```python
print(ANN)
```

```python
#2 definir la función de la Ecuación diferencial
F = lambda utt,uxx: utt - 3*uxx 
```

# ANN=ANN(t,x)

```python
#3 Se establece la función de coste
def coste(t):
    t.requires_grad=True
    #calculamos una salida de la red
    salida=ANN(t)
    #calculamos la derivada de la red respecto a esa entrada y salida, recordemos que cada elemento
    #de la función grad(), es la derivada parcial respecto a cada  una de las entradas
    DANN = torch.autograd.grad(salida, t, grad_outputs=torch.ones_like(salida),create_graph=True)[0] #Gradiente
    ANN_t=DANN[:,0] #dANN/dt
    ANN_x=DANN[:,1] #dANN/dx
    
    #Ahora calculamos las derivadas de segundo orden
    DDANN = torch.autograd.grad(DANN, t, grad_outputs=torch.ones_like(DANN),create_graph=True)[0]
    ANN_tt=DANN[:,0] #dANN/dt
    ANN_xx=DANN[:,1] #dANN/dx
    
    a=torch.cat((torch.zeros(t.shape[0],1),t[:,1].reshape(t.shape[0],1)),axis=1) #u(0,x)
    
    b=torch.cat((t[:,0].reshape(t.shape[0],1),torch.zeros(t.shape[0],1)),axis=1) #u(t,0)
    
    c=torch.cat((t[:,0].reshape(t.shape[0],1),5*torch.ones(t.shape[0],1)),axis=1) #u(t,L=5)
    

    return  torch.mean(F(ANN_tt,ANN_xx)** 2) + torch.mean(ANN(c)**2) + torch.mean(ANN(b)**2) + torch.mean((ANN(a)-torch.sin(t[:,1].reshape(t.shape[0],1)))**2)
```

```python
#5 el intervalo que queremos para x
x = torch.arange(0,5,0.1)
t = torch.arange(0,10,0.1)
X=torch.cartesian_prod(t,x)
X.requires_grad=True
```

```python
#loop de entrenamiendo
learning_rate=0.1
epocas=500

#4 El optimizador
optimizer=torch.optim.Adam(ANN.parameters(),lr=learning_rate)

errores=[]
for i in range(epocas):
    l=coste(X) #coste
    l.backward() #gradiente
    optimizer.step() #se actualizan los parámetros
    optimizer.zero_grad() #vacíamos el gradiente
    print('época=%.2f , loss=%.5f'%(i,l))
    errores.append(float(l))
plt.plot(range(epocas),errores)
```
